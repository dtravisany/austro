```python
import pandas as pd
import seaborn as sns
from scipy.spatial.distance import pdist,hamming,squareform
import pylab
import scipy.cluster.hierarchy  as sch
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

## Cargamos los datos de la matriz en un pandas Dataframe

```python
df = pd.read_excel("matrizCristina.xlsx",index_col="#reaction")
```

### Exploramos la forma de la transpuesta

```python
df.shape
```


## Generamos la transpuesta para tener un vector fila de reacciones por especie

```python
df.T
```

### Guardamos los nombres de las especies para utilizarlos como labels en las imágenes posteriores

```python
names=df.T.index
names
```

### Generamos el primer árbol para esto calculamos la distancia y graficamos 

```python
data=pdist(df.T,'hamming') # Distancia de Hamming, dado los valores booleanos, calcula la max distancia entre vectores.
D=data
D=squareform(D)
h,y=D.shape
fig = pylab.figure(figsize=(25,10))
ax1 = fig.add_axes([0.097,0.1,0.1,0.6],frameon=False)
Y = sch.linkage(D[np.triu_indices(h, 1)], method='complete') #clustering utilizando complete linkage y el metodo complete, que permite obtener la mayor separacion entre los clsters.  
Z1 = sch.dendrogram(Y,labels=names, orientation='left',leaf_font_size=12)
```

# Generamos el dibujo de los arboles superior y lateral izq, luego graficamos la matriz y agregamos la barra leyenda.

```python
#Calculamos el primer arbol

data
D=squareform(data)
h,y=D.shape
fig = pylab.figure(figsize=(20,20))
ax1 = fig.add_axes([0.09,0.1,0.2,0.6])
Y = sch.linkage(D[np.triu_indices(h, 1)], method='complete')
Z1 = sch.dendrogram(Y, orientation='left')
ax1.set_xticks([])
ax1.set_yticks([])

ax2 = fig.add_axes([0.25,0.71,0.6,0.2])
Y = sch.linkage(D[np.triu_indices(h, 1)], method='complete')
Z2 = sch.dendrogram(Y)
ax2.set_xticks([])
ax2.set_yticks([])




axmatrix = fig.add_axes([0.25,0.1,0.6,0.6])
idx1 = Z1['leaves']
idx2 = Z2['leaves']

D = D[idx1,:]
D = D[:,idx2]


stidx2=[]
for i in idx2:
    stidx2.append(names[i])
stidx1=[]
for i in idx1:
    stidx1.append(names[i])

im = axmatrix.matshow(D, aspect='auto', origin='lower', cmap=pylab.cm.Blues_r)


#Limpiamos los ticks
axmatrix.set_xticks([])
axmatrix.set_yticks([])
#Agregamos los nombres a los ticks en x
axmatrix.set_xticks(range(len(names)))
axmatrix.set_xticklabels(stidx1, minor=False,rotation=90,fontsize=18)
#Ponemos los labels de X en la abajo 
axmatrix.xaxis.set_label_position('bottom')
axmatrix.xaxis.tick_bottom()
#Agregamos los nombres de los ticks en y
axmatrix.set_yticks(range(len(names)))
axmatrix.set_yticklabels(stidx2,minor=False,fontsize=18)
#Ponemos los labels de Y en la derecha 
axmatrix.yaxis.set_label_position('right')
axmatrix.yaxis.tick_right()

axcolor = fig.add_axes([0.955,0.1,0.02,0.6])

#Agregamos la barra leyenda
pylab.colorbar(im, cax=axcolor)


fig.show()
#Guardamos la figura
plt.savefig("FiguraMatrix.png")
```
